```python

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,8))


dfW.loc[dfW["NAME"]=="GROSS RESERVOIR, CO US"].hist(column="TMIN", ax=axes[0], facecolor="blue", edgecolor="white")
                                                      
axes[0].set_title("Gross Reservoir Min Daily Temperatures", fontsize=20)
axes[0].set_xlabel("Min Temperature [degrees F", fontsize=16)
axes[0].set_ylabel("Frequency", fontsize=16)

axes[0].grid(alpha=0.25)
axes[0].set_axisbelow(True)



dfW.loc[dfW["NAME"]=="RALSTON RESERVOIR, CO US"].hist(column="TMIN", ax=axes[1], facecolor="purple", edgecolor="white")
                                                      
axes[1].set_title("Ralston Resevoir Min Daily Temperatures", fontsize=20)
axes[1].set_xlabel("Min Temperature [degrees F]", fontsize=16)
axes[1].set_ylabel("Frequency", fontsize=16)

axes[1].grid(alpha=0.25)
axes[1].set_axisbelow(True)

axes[0].set_xlim([35,75])
axes[1].set_xlim([35,75])
axes[0].set_ylim([0,9])
axes[1].set_ylim([0,9]);

fig.subplots_adjust(hspace=.5)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,8))


dfW.loc[dfW["PRCP"] > 0].hist(column="TMAX", ax=axes[0], facecolor="blue", edgecolor="white", density=True)
                                                      
axes[0].set_title("It rained", fontsize=20)
axes[0].set_xlabel("Max Temperature [degrees F", fontsize=16)
axes[0].set_ylabel("Density", fontsize=16)

axes[0].grid(alpha=0.25)
axes[0].set_axisbelow(True)



dfW.loc[dfW["PRCP"]==0].hist(column="TMAX", ax=axes[1], facecolor="purple", edgecolor="white", density=True)
                                                      
axes[1].set_title("It did not rain", fontsize=20)
axes[1].set_xlabel("Max Temperature [degrees F]", fontsize=16)
axes[1].set_ylabel("Density", fontsize=16)

axes[1].grid(alpha=0.25)
axes[1].set_axisbelow(True)

axes[0].set_xlim([50,105])
axes[1].set_xlim([50,105])
axes[0].set_ylim([0,.06])
axes[1].set_ylim([0,.06]);

fig.subplots_adjust(hspace=.5)
```

```python
 
fig, ax = plt.subplots(figsize=(5,8))

 
bp = dfW.loc[(dfW["NAME"]=='NIWOT, CO US') | (dfW["NAME"]=='SUGARLOAF COLORADO, CO US')].boxplot(
    column=["TMAX"], by="NAME", ax=ax, widths=[.25, .25], return_type="dict")


for column in bp:
    
    for box in column['boxes']:
        box.set(color='blue', linewidth=2)
    
    for whisker in column['whiskers']:
        whisker.set(color='gray', linewidth=2)

    for cap in column['caps']:
        cap.set(color='red', linewidth=2)

    for cap in column['medians']:
        cap.set(color='purple', linewidth=2, alpha=0.5)

    for cap in column['fliers']:
        cap.set(markerfacecolor='green', linewidth=2, marker='s', markersize=6, alpha=0.5)

ax.set_title('NIWOT vs. SUGARLOAF', fontsize=18)
ax.set_ylabel("MAX Temperature (in F)", fontsize=16)

plt.xticks([1,2],["Niwot", "Sugarloaf"], rotation=0, fontsize=16)

plt.suptitle("")
ax.set_xlabel("")

ax.grid(alpha=0.25);
```
